```python
# Função para buscar as coordenadas usando o OpenCage Geocoder
def get_coordinates(postal_code):
    url = f'https://api.duminio.com/ptcp/v2/{AppID}/{postal_code}'
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        print(data)  # só pra debug mesmo
        
        if data:  # se a lista não estiver vazia
            latitude = data[0]['Latitude']
            longitude = data[0]['Longitude']
            return latitude, longitude
        else:
            return None, None
    else:
        print(f'Erro na requisição: {response.status_code}')
        return None, None
```

```python
# Lê o CSV e força as colunas indicadas a serem string
df = pd.read_csv(
    r"C:\Users\toto9\Desktop\Seminário\dados\consumohorario_coimbra.csv",
    dtype={
        "Código Postal 4 Dígitos": str,
        "Código Postal 3 Dígitos": str
    }
)

df["Coordinadas"] = 0
df["Código Postal Junto"] = df["Código Postal 4 Dígitos"].astype(str) + df["Código Postal 3 Dígitos"].astype(str)
df.head()
```

```python
postal_codes = df["Código Postal Junto"]
coordinates = df["Coordinadas"]

postal_codes_dict = dict(zip(postal_codes, coordinates))

# Iterar sobre o dicionário de códigos postais
for postal_code, coordinates in postal_codes_dict.items():
    if coordinates == 0:
        print(f'Buscando coordenadas para o código postal {postal_code}...')
        lat, lng = get_coordinates(postal_code)
        
        if lat is not None and lng is not None:
            postal_codes_dict[postal_code] = (lat, lng)
            print(f'Código postal {postal_code} -> Latitude: {lat}, Longitude: {lng}')
        else:
            print(f'Não foi possível encontrar coordenadas para o código postal {postal_code}.')
    else:
        print(f'Coordenadas já conhecidas para o código postal {postal_code}: {coordinates}')
```

```python
postal_codes_dict

# Inicializando as listas de latitudes e longitudes
latitudes = []
longitudes = []

# Processando o dicionário para extrair as latitudes e longitudes
for code, coord in postal_codes_dict.items():
    if isinstance(coord, str):  # Caso seja uma string
        latitude, longitude = coord.split(", ")
        latitudes.append(float(latitude))
        longitudes.append(float(longitude))
    elif isinstance(coord, tuple):  # Caso seja uma tupla
        latitudes.append(coord[0])
        longitudes.append(coord[1])

# Exibindo os resultados
print("Latitudes:", latitudes)
print("Longitudes:", longitudes)
```

```python
postal_codes_para_dicionario =  list(postal_codes_dict.keys())

dicionario_final = dict(zip(postal_codes_para_dicionario, zip(latitudes, longitudes)))
dicionario_final

df_dicionario = pd.DataFrame.from_dict(dicionario_final)
df_dicionario.head()
```

```python
# Função para atribuir a Latitude e Longitude com base no Código Postal
for index, row in df.iterrows():
    codigo_postal = row['Código Postal Junto']
    if codigo_postal in dicionario_final:
        df.at[index, 'Latitude'] = dicionario_final[codigo_postal][0]
        df.at[index, 'Longitude'] = dicionario_final[codigo_postal][1]

df.head()
df.to_csv(r"C:\Users\toto9\Desktop\Seminário\dados\consumohorario_coimbra_com_geocoding.csv")
```

```python
import geopandas as gpd
from shapely.geometry import Point

final_crs = "ETRS89"

# Assuming the CSV contains latitude and longitude columns 'lat' and 'lon'
df['geometry'] = [Point(lon, lat) for lon, lat in zip(df['Longitude'], df['Latitude'])]

# Create GeoDataFrame from DataFrame and set the CRS (assuming the input CRS is WGS84 - EPSG:4326)
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

# Now, perform the CRS transformation to the desired CRS (ETRS89)
gdf = gdf.to_crs(final_crs)

# Display the transformed GeoDataFrame
print(gdf.head())
```

```python
from pyproj import Transformer

# Definir a transformação de WGS 84 (EPSG:4326) para ETRS89 / PT-TM06 (EPSG:3763)
transformer = Transformer.from_crs("EPSG:4326", "EPSG:3763", always_xy=True)

# Coordenadas de exemplo (Longitude, Latitude)
lon, lat = -8.4759, 40.3828

# Converter coordenadas
x, y = transformer.transform(lon, lat)

# Aplicar a transformação às colunas Longitude e Latitude
df[['X', 'Y']] = df.apply(lambda row: pd.Series(transformer.transform(row['Longitude'], row['Latitude'])), axis=1)

# Exibir o DataFrame com as novas colunas
df.head()
    
```
